### 3-2 데이터 분석하기

#### 필수 문제1 : 월별 매출액 추세(sales_data를 이용합니다.)
월별 매출액 추세를 집계하고 그 결과를 시각화해봅니다.

힌트: groupby() 함수를 사용하세요.

```python
# 활용할 데이터(sales_data)를 temp_var로 정의하기
sales_data['결제일시'] = sales_data['결제일시'].apply(lambda x : str(x[:4])+str(x[5:7])).astype(int)
temp_var = sales_data.groupby('결제일시').size()
temp_var = temp_var.to_frame().reset_index()
temp_var.columns = ['결제년월', 'count']
temp_var.insert(2, '월별매출액', 0)

# 결제금액에서 콤마(,)를 제거하고 월별매출액에 합산하기
for i in range(len(sales_data)):
    for k in range(len(temp_var)):
        if sales_data['결제일시'][i] == temp_var['결제년월'][k]:
            temp_var['월별매출액'][k] += int(sales_data['결제금액'][i].replace(',',''))
temp_var
```

```python
# 한글폰트 불러오기
import matplotlib.font_manager as fm

font_list = [font.name for font in fm.fontManager.ttflist]
print(font_list)
plt.rcParams['font.family'] = 'NanumGothic'

# Bar Chart 그리기
ax = sns.barplot(data=temp_var, x='결제년월', y='월별매출액')
ax.set_title('월별 음식/음료 매출액', fontsize=15)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

#### 필수 문제2 : 가장 인기 있는 음식/음료(salses_data를 이용합니다.)
여러 메뉴들(음식,음료)을 판매 개수 순으로 정렬하고 상위 10개를 뽑아 결과를 시각화해봅니다.

힌트: value_counts() 함수를 사용하세요.

```python
# temp_var의 상품명을 value_counts를 이용해 집계하기
# 주의! value_counts로 바로 집계하면 판매취소와 결제취소가 함께 집계됩니다.
# 따라서 '판매수량' 칼럼 값이 '-1'인 데이터를 제외하고 집계해야 합니다.
# 힌트 코드 : temp_var[temp_var['판매수량']==-1]
temp_var = sales_data['상품명'].value_counts()
temp_var = temp_var.to_frame().reset_index()
temp_var.columns = ['상품명', 'count']
temp_var.insert(2, '실 판매건수', 0)

for i in range(len(sales_data)):
    for k in range(len(temp_var)):
        if sales_data['상품명'][i] == temp_var['상품명'][k]:
            temp_var['실 판매건수'][k] += int(sales_data['판매수량'][i])
temp_var
```

```python
# 집계 결과를 데이터프레임으로 변환하고(위에서 이미 변환했음), 상위 10개 선정하기
temp_var = temp_var.sort_values(by=['실 판매건수'], ascending=False)
temp_var = temp_var.drop(temp_var.index[10:])
temp_var
```

```python
# 한글이 포함된 시각화를 위해 Matplotlib 설정
import matplotlib.font_manager as fm

font_list = [font.name for font in fm.fontManager.ttflist]
print(font_list)
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
# 집계 결과를 시각화하기
ax = sns.barplot(data=temp_var, x='상품명', y='실 판매건수')
ax.set_title('상품별 판매건수', fontsize=15)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

여러 메뉴들(음식,음료)을 판매 매출(판매 개수 x 결제 금액) 순으로 정렬하고 상위 10개를 뽑아 결과를 시각화해봅니다.

힌트: value_counts() 함수를 사용하세요.

```python
# sales_data의 '결제금액'열을 '상품명'을 기준으로 집계하기
temp_var = sales_data['상품명'].value_counts()
temp_var = temp_var.to_frame().reset_index()
temp_var.columns = ['상품명', 'count']
temp_var.insert(2, '실 판매금액', 0)

for i in range(len(sales_data)):
    for k in range(len(temp_var)):
        if sales_data['상품명'][i] == temp_var['상품명'][k]:
            temp_var['실 판매금액'][k] += int(sales_data['결제금액'][i].replace(',',''))
temp_var
```

```python
# '실 판매금액'의 값을 내림차순으로 정렬하고 상위 10개 도출하기
temp_var = temp_var.sort_values(by=['실 판매금액'], ascending=False)
temp_var = temp_var.drop(temp_var.index[10:])
temp_var
```

```python
# 집계 결과 시각화하기
ax = sns.barplot(data=temp_var, x='상품명', y='실 판매금액')
ax.set_title('상품별 실 판매금액', fontsize=15)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

#### 필수 문제3: 가장 많이 팔린 입장권 종류(entry_data를 이용합니다.)

### 내가 정의한 문제 해결하기

#### 월별 입장권 매출액 추세를 집계하고 시각화

```python
## 월별 입장권 매출액을 집계하고 시각화 ##

# 활용할 데이터(entry_data)를 temp_var로 정의하기
entry_data['입실'] = entry_data['입실'].apply(lambda x : str(x[:4])+str(x[5:7])).astype(int)
temp_var = entry_data.groupby('입실').size()
temp_var = temp_var.to_frame().reset_index()
temp_var.columns = ['결제년월', 'count']
temp_var.insert(2, '월별매출액', 0)

# 금액을 숫자로 바꾸고 월별매출액에 합산하기
for i in range(len(entry_data)):
    for k in range(len(temp_var)):
        if entry_data['입실'][i] == temp_var['결제년월'][k]:
            temp_var['월별매출액'][k] += entry_data['금액'][i]
temp_var
```

```python
# 한글폰트 불러오기
import matplotlib.font_manager as fm

font_list = [font.name for font in fm.fontManager.ttflist]
print(font_list)
plt.rcParams['font.family'] = 'NanumGothic'

# Bar Chart 그리기
ax = sns.barplot(data=temp_var, x='결제년월', y='월별매출액')
ax.set_title('월별 입장권 매출액', fontsize=15)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```
